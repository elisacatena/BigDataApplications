```python
import pandas as pd
```

```python
def conta_righe(file_path):
    with open(file_path, 'r') as file:
        numero_righe = len(file.readlines())
    return numero_righe
```

```python
file_path_1 = "merged_data_1.csv"  # Inserisci il percorso del tuo file qui
file_1 = pd.read_csv(file_path_1)
numero_di_righe_1 = conta_righe(file_path_1)
print("Il file ha", numero_di_righe_1, "righe.")
file_1
```
