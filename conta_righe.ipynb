```python
import pandas as pd
```

```python
def conta_righe(file_path):
    with open(file_path, 'r') as file:
        numero_righe = len(file.readlines())
    return numero_righe
```

```python
file_path = "input/merged_data.csv" 
file = pd.read_csv(file_path)
numero_di_righe = conta_righe(file_path)
print("Il file ha", numero_di_righe, "righe.")
file
```
