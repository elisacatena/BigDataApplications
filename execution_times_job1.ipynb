```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Tempi di esecuzione in locale

```python
job_data_1 = [
    {"input_size": 2.81, "execution_time": 5.20},
    {"input_size": 3.40, "execution_time": 7.07},
    {"input_size": 3.99, "execution_time": 9.17},
    {"input_size": 4.58, "execution_time": 11},
]

df1 = pd.DataFrame(job_data_1)
fig, ax = plt.subplots(figsize=(10, 5), facecolor='white')
ax.set_facecolor('white')

ax.plot(df1['input_size'], df1['execution_time'], marker='o', linestyle='-', color='mediumaquamarine', markersize=8, linewidth=2)

# Aggiungere il bordo nero
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Etichette degli assi
ax.set_xlabel("Dimensione dell'input (GB)", fontsize=12, labelpad=15)
ax.set_ylabel("Tempo di esecuzione (min)", fontsize=12, labelpad=15)

# Titolo del grafico
ax.set_title("Tempo di esecuzione MapReduce (in locale) job1 in funzione della dimensione dell'input", fontsize=16, pad=20)

ax.grid(True, linestyle='--', linewidth=0.5, color='lightgray')

plt.show()
```
