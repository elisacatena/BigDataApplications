```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
job_data = [
    {"input_size": 2.81, "execution_time": 7.16},
    {"input_size": 3.40, "execution_time": 9.04},
    {"input_size": 3.99, "execution_time": 11.22},
    {"input_size": 4.58, "execution_time": 13.45},
]

# Creare un DataFrame
df = pd.DataFrame(job_data)

# Creare il grafico
fig, ax = plt.subplots(figsize=(10, 5), facecolor='white')
# Impostare lo sfondo interno del grafico come bianco
ax.set_facecolor('white')

# Personalizzare il grafico
ax.plot(df['input_size'], df['execution_time'], marker='o', linestyle='-', color='mediumaquamarine', markersize=8, linewidth=2)

# Aggiungere il bordo nero
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)

# Etichette degli assi
ax.set_xlabel("Dimensione dell'input (GB)", fontsize=12, labelpad=15)
ax.set_ylabel("Tempo di esecuzione (min)", fontsize=12, labelpad=15)

# Titolo del grafico
ax.set_title("Tempo di esecuzione MapReduce (in locale) job1 in funzione della dimensione dell'input", fontsize=14, pad=20)

# Aggiungere una griglia sottile
ax.grid(True, linestyle='--', linewidth=0.5, color='lightgray')

plt.show()
```
