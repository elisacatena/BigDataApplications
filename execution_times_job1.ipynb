```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
job_data = [
    {"input_size": 2.81, "execution_time": 7},
    {"input_size": 3.40, "execution_time": 7},
    {"input_size": 3.99, "execution_time": 9},
    {"input_size": 4.58, "execution_time": 11},
]

# Creare un DataFrame
df = pd.DataFrame(job_data)

# Creare il grafico
plt.figure(figsize=(10, 5))
plt.plot(df['input_size'], df['execution_time'], marker='o', linestyle='-', color='mediumaquamarine')
plt.xlabel('Dimensione dell\'input (GB)')
plt.ylabel('Tempo di esecuzione (min)')
plt.title('Tempo di esecuzione MapReduce (in locale) job1 in funzione della dimensione dell\'input')
plt.grid(True)
plt.show()
```
