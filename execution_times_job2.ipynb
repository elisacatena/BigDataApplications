```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
job_data = [
    {"input_size": 2.81, "execution_time": 4},
    {"input_size": 3.40, "execution_time": 4},
    {"input_size": 3.99, "execution_time": 7.18},
    {"input_size": 4.58, "execution_time": 8.48},
]

# Creare un DataFrame
df = pd.DataFrame(job_data)

# Creare il grafico
plt.figure(figsize=(10, 5))
plt.plot(df['input_size'], df['execution_time'], marker='o', linestyle='-', color='pink')
plt.xlabel('Dimensione dell\'input (GB)')
plt.ylabel('Tempo di esecuzione (min)')
plt.title('Tempo di esecuzione MapReduce (in locale) job2 in funzione della dimensione dell\'input')
plt.grid(True)
plt.show()
```
